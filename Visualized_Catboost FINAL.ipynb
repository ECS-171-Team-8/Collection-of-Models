```python
# Part 1: Imports & basic setup
!pip install catboost --quiet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (
    accuracy_score, f1_score, classification_report,
    confusion_matrix, roc_auc_score, roc_curve, auc
)
from sklearn.preprocessing import label_binarize

from catboost import CatBoostClassifier, Pool

RANDOM_STATE = 42
TARGET_COL = "NObeyesdad"
N_FOLDS = 5

plt.style.use("ggplot")
pd.set_option("display.max_columns", 100)
pd.set_option("display.width", 120)

# Part 2: Load data
train_path = "sample_data/train.csv"
test_path = "sample_data/test.csv"
sub_path = "sample_data/sample_submission.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
sample_submission = pd.read_csv(sub_path)

print("Train shape:", train_df.shape)
print("Test shape :", test_df.shape)
print("\nColumns in train:")
print(train_df.columns.tolist())

# Part 3: Basic preprocessing
y = train_df[TARGET_COL]
X = train_df.drop(columns=[TARGET_COL])

if "id" in X.columns:
    X = X.drop(columns=["id"])
    print("Dropped 'id' from features.")

test_ids = None
if "id" in test_df.columns:
    test_ids = test_df["id"].copy()
    test_df = test_df.drop(columns=["id"])
    print("Dropped 'id' from test features and stored as test_ids.")

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()

print("\nNumeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
print("\nTarget distribution:")
print(y.value_counts())

cat_feature_indices = [X.columns.get_loc(col) for col in categorical_cols]
print("\nCategorical feature indices (CatBoost):", cat_feature_indices)

# ==========================================
# Part 4: Hyperparameter Tuning
# ==========================================
print(f"\n{'='*60}")
print("Hyperparameter Tuning")
print(f"{'='*60}")

# Split data for tuning
X_train_tune, X_val_tune, y_train_tune, y_val_tune = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print(f"Tuning train size: {X_train_tune.shape[0]}")
print(f"Tuning validation size: {X_val_tune.shape[0]}")

# Define parameter grid
param_grid = {
    "learning_rate": [0.05, 0.1, 0.15],
    "depth": [4, 6, 8],
    "l2_leaf_reg": [1.0, 3.0, 5.0]
}

# Store tuning results
results_list = []

# Grid search
for lr in param_grid["learning_rate"]:
    for depth in param_grid["depth"]:
        for l2_reg in param_grid["l2_leaf_reg"]:
            print(f"\n=== Training CatBoost: lr={lr}, depth={depth}, l2_reg={l2_reg} ===")
            
            # Create pools for tuning
            train_pool_tune = Pool(
                data=X_train_tune,
                label=y_train_tune,
                cat_features=cat_feature_indices
            )
            
            val_pool_tune = Pool(
                data=X_val_tune,
                label=y_val_tune,
                cat_features=cat_feature_indices
            )
            
            # Create model with current parameters
            model_tmp = CatBoostClassifier(
                loss_function="MultiClass",
                eval_metric="TotalF1",
                learning_rate=lr,
                depth=depth,
                l2_leaf_reg=l2_reg,
                random_seed=RANDOM_STATE,
                iterations=500,
                early_stopping_rounds=50,
                verbose=False
            )
            
            # Train model
            model_tmp.fit(
                train_pool_tune,
                eval_set=val_pool_tune,
                use_best_model=True
            )
            
            # Predictions
            y_pred = model_tmp.predict(val_pool_tune).reshape(-1)
            y_pred_proba = model_tmp.predict_proba(val_pool_tune)
            
            # Calculate metrics
            acc_tmp = accuracy_score(y_val_tune, y_pred)
            f1_tmp = f1_score(y_val_tune, y_pred, average="macro")
            auc_tmp = roc_auc_score(
                label_binarize(y_val_tune, classes=model_tmp.classes_),
                y_pred_proba,
                multi_class="ovr",
                average="macro"
            )
            
            results_list.append({
                "learning_rate": lr,
                "depth": depth,
                "l2_leaf_reg": l2_reg,
                "accuracy": acc_tmp,
                "macro_f1": f1_tmp,
                "auc": auc_tmp,
                "best_iteration": model_tmp.get_best_iteration()
            })
            
            print(f"Accuracy: {acc_tmp:.4f}, Macro-F1: {f1_tmp:.4f}, AUC: {auc_tmp:.4f}")

# Create results DataFrame
results_df = pd.DataFrame(results_list)
print("\n" + "="*60)
print("Parameter Tuning Results (sorted by AUC)")
print("="*60)
print(results_df.sort_values("auc", ascending=False).to_string(index=False))

# Find best parameters
best_params = results_df.loc[results_df["auc"].idxmax()]
print(f"\n{'='*60}")
print("Best Parameters:")
print(f"{'='*60}")
print(f"Learning Rate: {best_params['learning_rate']}")
print(f"Depth: {int(best_params['depth'])}")
print(f"L2 Leaf Reg: {best_params['l2_leaf_reg']}")
print(f"Best Accuracy: {best_params['accuracy']:.4f}")
print(f"Best Macro-F1: {best_params['macro_f1']:.4f}")
print(f"Best AUC: {best_params['auc']:.4f}")

# Visualization 1: Performance vs Parameters (Original Style)
plt.figure(figsize=(10, 6))
plt.plot(results_df["auc"], marker="o", label="AUC")
plt.plot(results_df["accuracy"], marker="s", label="Accuracy")
plt.xticks(ticks=range(len(results_df)),
           labels=[f"lr={row['learning_rate']}, md={int(row['depth'])}"
                   for _, row in results_df.iterrows()],
           rotation=45)
plt.title("Performance vs Parameters")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

# Visualization 2: Top 10 Parameter Configurations
top_10 = results_df.nlargest(10, 'auc').copy()
top_10['config'] = top_10.apply(
    lambda x: f"lr={x['learning_rate']}, d={int(x['depth'])}, l2={x['l2_leaf_reg']}", 
    axis=1
)

plt.figure(figsize=(12, 6))
x_pos = np.arange(len(top_10))
width = 0.25

plt.bar(x_pos - width, top_10['auc'], width, label='AUC', color='steelblue', alpha=0.8)
plt.bar(x_pos, top_10['accuracy'], width, label='Accuracy', color='coral', alpha=0.8)
plt.bar(x_pos + width, top_10['macro_f1'], width, label='Macro-F1', color='seagreen', alpha=0.8)

plt.xlabel('Parameter Configuration', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Top 10 Parameter Configurations', fontsize=14, fontweight='bold')
plt.xticks(x_pos, top_10['config'], rotation=45, ha='right', fontsize=8)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Visualization 3: Heatmap - AUC by Learning Rate and Depth
pivot_data = results_df.groupby(['learning_rate', 'depth'])['auc'].mean().reset_index()
pivot_table = pivot_data.pivot(index='depth', columns='learning_rate', values='auc')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='.4f', cmap='YlOrRd', 
            cbar_kws={'label': 'AUC Score'})
plt.title('Average AUC by Learning Rate and Depth', fontsize=14, fontweight='bold')
plt.xlabel('Learning Rate', fontsize=12)
plt.ylabel('Depth', fontsize=12)
plt.tight_layout()
plt.show()

# Update hyperparameters for cross-validation
BEST_LR = best_params['learning_rate']
BEST_DEPTH = int(best_params['depth'])
BEST_L2_REG = best_params['l2_leaf_reg']

print(f"\n{'='*60}")
print(f"Using best parameters for {N_FOLDS}-Fold Cross Validation:")
print(f"Learning Rate: {BEST_LR}, Depth: {BEST_DEPTH}, L2 Reg: {BEST_L2_REG}")
print(f"{'='*60}")

# ==========================================
# Part 5: K-Fold Cross Validation with Best Parameters
# ==========================================
print(f"\n{'='*60}")
print(f"Starting {N_FOLDS}-Fold Cross Validation")
print(f"{'='*60}")

# Initialize cross-validation
skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)

# Store results for each fold
fold_scores = []
fold_models = []
oof_predictions = np.empty(len(X), dtype=object)  # Out-of-fold predictions (string type)
oof_probas = np.zeros((len(X), len(y.unique())))
test_predictions = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y), 1):
    print(f"\n{'='*60}")
    print(f"Fold {fold}/{N_FOLDS}")
    print(f"{'='*60}")
    
    # Split training and validation sets
    X_train_fold = X.iloc[train_idx]
    y_train_fold = y.iloc[train_idx]
    X_val_fold = X.iloc[val_idx]
    y_val_fold = y.iloc[val_idx]
    
    print(f"Train size: {len(train_idx)}, Validation size: {len(val_idx)}")
    
    # Create CatBoost Pool
    train_pool = Pool(
        data=X_train_fold,
        label=y_train_fold,
        cat_features=cat_feature_indices
    )
    
    val_pool = Pool(
        data=X_val_fold,
        label=y_val_fold,
        cat_features=cat_feature_indices
    )
    
    # Train model
    model = CatBoostClassifier(
        loss_function="MultiClass",
        eval_metric="TotalF1",
        learning_rate=BEST_LR,
        depth=BEST_DEPTH,
        l2_leaf_reg=BEST_L2_REG,
        random_seed=RANDOM_STATE,
        iterations=1000,
        early_stopping_rounds=100,
        verbose=100
    )
    
    model.fit(
        train_pool,
        eval_set=val_pool,
        use_best_model=True
    )
    
    # Validation set predictions
    y_val_pred = model.predict(val_pool).reshape(-1)
    y_val_proba = model.predict_proba(val_pool)
    
    # Save out-of-fold predictions
    oof_predictions[val_idx] = y_val_pred
    oof_probas[val_idx] = y_val_proba
    
    # Calculate metrics
    fold_acc = accuracy_score(y_val_fold, y_val_pred)
    fold_f1 = f1_score(y_val_fold, y_val_pred, average="macro")
    
    fold_scores.append({
        'fold': fold,
        'accuracy': fold_acc,
        'macro_f1': fold_f1,
        'best_iteration': model.get_best_iteration()
    })
    
    print(f"\nFold {fold} Results:")
    print(f"  Accuracy  : {fold_acc:.4f}")
    print(f"  Macro-F1  : {fold_f1:.4f}")
    print(f"  Best iter : {model.get_best_iteration()}")
    
    # Save model
    fold_models.append(model)
    
    # Test set predictions (for subsequent ensemble)
    test_pool = Pool(
        data=test_df,
        cat_features=[test_df.columns.get_loc(col) for col in categorical_cols]
    )
    test_pred_proba = model.predict_proba(test_pool)
    test_predictions.append(test_pred_proba)

# ==========================================
# Visualization 1: Cross-Validation Scores
# ==========================================
def plot_cv_scores(fold_scores):
    """Plot cross-validation scores"""
    scores_df = pd.DataFrame(fold_scores)
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Accuracy
    axes[0].bar(scores_df['fold'], scores_df['accuracy'], 
                color='steelblue', alpha=0.7)
    axes[0].axhline(scores_df['accuracy'].mean(), color='red', 
                    linestyle='--', linewidth=2, 
                    label=f"Mean: {scores_df['accuracy'].mean():.4f}")
    axes[0].set_xlabel('Fold', fontsize=12)
    axes[0].set_ylabel('Accuracy', fontsize=12)
    axes[0].set_title('Cross-Validation Accuracy', fontsize=14, fontweight='bold')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Macro F1
    axes[1].bar(scores_df['fold'], scores_df['macro_f1'], 
                color='seagreen', alpha=0.7)
    axes[1].axhline(scores_df['macro_f1'].mean(), color='red', 
                    linestyle='--', linewidth=2,
                    label=f"Mean: {scores_df['macro_f1'].mean():.4f}")
    axes[1].set_xlabel('Fold', fontsize=12)
    axes[1].set_ylabel('Macro F1', fontsize=12)
    axes[1].set_title('Cross-Validation Macro F1', fontsize=14, fontweight='bold')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

plot_cv_scores(fold_scores)

# ==========================================
# Visualization: Training Curve (First Fold)
# ==========================================
def plot_training_curve(model, fold_num=1):
    """Plot training and validation curves for a specific fold"""
    # Get training history
    results = model.get_evals_result()
    
    # Extract training and validation metrics
    train_metric = results['learn']['TotalF1']
    val_metric = results['validation']['TotalF1']
    
    # Find best score
    best_score = max(val_metric)
    best_epoch = val_metric.index(best_score)
    
    plt.figure(figsize=(8, 5))
    plt.plot(train_metric, label='Train')
    plt.plot(val_metric, label='Validation')
    plt.scatter(best_epoch, best_score, color='green', label='Best')
    plt.xlabel('Epoch')
    plt.ylabel('TotalF1 Score')
    plt.title('Training Curve')
    plt.legend()
    plt.show()
    
    print(f"Train F1 (final): {train_metric[-1]:.4f}")
    print(f"Val F1 (final):   {val_metric[-1]:.4f}")

# Plot training curve for the first fold
print(f"\n{'='*60}")
print("Training Curve Analysis (Fold 1)")
print(f"{'='*60}")
plot_training_curve(fold_models[0], fold_num=1)

# ==========================================
# Visualization: MultiClass Log Loss Curve (First Fold)
# ==========================================
def plot_logloss_curve(model, fold_num=1):
    """Plot training and validation log loss curves for a specific fold"""
    # Get training history
    results = model.get_evals_result()
    
    # Extract training and validation log loss
    train_loss = results['learn']['MultiClass']
    val_loss = results['validation']['MultiClass']
    
    # Find best loss
    best_loss = min(val_loss)
    best_epoch = val_loss.index(best_loss)
    
    plt.figure(figsize=(8, 5))
    plt.plot(train_loss, label='Train')
    plt.plot(val_loss, label='Validation')
    plt.scatter(best_epoch, best_loss, color='green', label='Best')
    plt.xlabel('Epoch')
    plt.ylabel('Multiclass Log Loss')
    plt.title('Training Curve')
    plt.legend()
    plt.show()
    
    print(f"Train Loss (final): {train_loss[-1]:.4f}")
    print(f"Val Loss (final):   {val_loss[-1]:.4f}")

# Plot log loss curve for the first fold
print(f"\n{'='*60}")
print("MultiClass Log Loss Analysis (Fold 1)")
print(f"{'='*60}")
plot_logloss_curve(fold_models[0], fold_num=1)

# Print cross-validation statistics
print(f"\n{'='*60}")
print("Cross-Validation Summary")
print(f"{'='*60}")
scores_df = pd.DataFrame(fold_scores)
print(scores_df.to_string(index=False))
print(f"\nMean Accuracy  : {scores_df['accuracy'].mean():.4f}  {scores_df['accuracy'].std():.4f}")
print(f"Mean Macro-F1  : {scores_df['macro_f1'].mean():.4f}  {scores_df['macro_f1'].std():.4f}")
print(f"Mean Best Iter : {scores_df['best_iteration'].mean():.1f}")

# ==========================================
# Part 5: Out-of-Fold Evaluation
# ==========================================
print(f"\n{'='*60}")
print("Out-of-Fold Evaluation")
print(f"{'='*60}")

oof_acc = accuracy_score(y, oof_predictions)
oof_f1 = f1_score(y, oof_predictions, average="macro")

print(f"\nOOF Accuracy  : {oof_acc:.4f}")
print(f"OOF Macro-F1  : {oof_f1:.4f}")

print("\nOOF Classification Report:")
print(classification_report(y, oof_predictions))

# ==========================================
# Visualization 2: OOF Confusion Matrix
# ==========================================
def plot_confusion_matrix(y_true, y_pred, classes, title="Confusion Matrix"):
    """Plot confusion matrix heatmap"""
    cm = confusion_matrix(y_true, y_pred, labels=classes)
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=classes, yticklabels=classes,
                cbar_kws={'label': 'Count'})
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(y, oof_predictions, fold_models[0].classes_, 
                     title="Out-of-Fold Confusion Matrix")

# ==========================================
# Visualization 3: OOF ROC Curves
# ==========================================
def plot_roc_curves_multiclass(y_true, y_proba, classes):
    """Plot multi-class ROC curves"""
    y_true_bin = label_binarize(y_true, classes=classes)
    n_classes = len(classes)
    
    plt.figure(figsize=(12, 8))
    
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2,
                label=f'{classes[i]} (AUC={roc_auc:.3f})')
    
    plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('OOF ROC Curve - Multi-class One-vs-Rest', 
              fontsize=14, fontweight='bold')
    plt.legend(loc='lower right', fontsize=9)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    macro_auc = roc_auc_score(y_true_bin, y_proba,
                              multi_class="ovr", average="macro")
    micro_auc = roc_auc_score(y_true_bin, y_proba,
                              multi_class="ovr", average="micro")
    
    print(f"\nOOF Macro AUC: {macro_auc:.4f}")
    print(f"OOF Micro AUC: {micro_auc:.4f}")

plot_roc_curves_multiclass(y, oof_probas, fold_models[0].classes_)

# ==========================================
# Visualization 4: Feature Importance (Average)
# ==========================================
def plot_feature_importance_cv(fold_models, feature_names, top_n=15):
    """Plot cross-validation average feature importance"""
    # Get feature importance from all folds
    all_importances = []
    for model in fold_models:
        all_importances.append(model.get_feature_importance())
    
    # Calculate mean and standard deviation
    mean_importance = np.mean(all_importances, axis=0)
    std_importance = np.std(all_importances, axis=0)
    
    # Create DataFrame
    fi_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': mean_importance,
        'Std': std_importance
    }).sort_values('Importance', ascending=False)
    
    # Display only top_n features
    fi_df_top = fi_df.head(top_n)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(
        x='Importance',
        y='Feature',
        data=fi_df_top,
        color='steelblue',
        alpha=0.7
    )
    plt.xlabel('Importance', fontsize=12)
    plt.ylabel('Feature', fontsize=12)
    plt.title(f'Top {top_n} Feature Importances - CatBoost (CV Average)', 
              fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    print(f"\nTop {top_n} most important features (CV average):")
    print(fi_df_top.to_string(index=False))
    
    return fi_df

fi_df = plot_feature_importance_cv(fold_models, X.columns, top_n=15)

# ==========================================
# Visualization 5: Cross-Validated Validation Accuracy
# ==========================================
print(f"\n{'='*60}")
print("Cross-Validated Validation Accuracy Analysis")
print(f"{'='*60}")

# Calculate cross-validated accuracy using OOF predictions
cv_val_acc = accuracy_score(y, oof_predictions)
print(f"\nCross-validated validation accuracy: {cv_val_acc:.4f}")

# Detailed classification report
print("\nDetailed Classification Report (Cross-Validated):")
print(classification_report(y, oof_predictions))

# Visualization: per-class accuracy
from sklearn.metrics import precision_recall_fscore_support

precision, recall, f1, support = precision_recall_fscore_support(
    y, oof_predictions, labels=fold_models[0].classes_
)

metrics_df = pd.DataFrame({
    'Class': fold_models[0].classes_,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1,
    'Support': support
})

print("\nPer-Class Metrics:")
print(metrics_df.to_string(index=False))

# Visualize per-class performance
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Precision
axes[0].barh(metrics_df['Class'], metrics_df['Precision'], color='coral', alpha=0.7)
axes[0].set_xlabel('Precision', fontsize=12)
axes[0].set_ylabel('Obesity Class', fontsize=12)
axes[0].set_title('Precision by Class (Cross-Validated)', fontsize=14, fontweight='bold')
axes[0].set_xlim([0, 1])
axes[0].grid(True, alpha=0.3)

# Recall
axes[1].barh(metrics_df['Class'], metrics_df['Recall'], color='lightgreen', alpha=0.7)
axes[1].set_xlabel('Recall', fontsize=12)
axes[1].set_ylabel('Obesity Class', fontsize=12)
axes[1].set_title('Recall by Class (Cross-Validated)', fontsize=14, fontweight='bold')
axes[1].set_xlim([0, 1])
axes[1].grid(True, alpha=0.3)

# F1-Score
axes[2].barh(metrics_df['Class'], metrics_df['F1-Score'], color='skyblue', alpha=0.7)
axes[2].set_xlabel('F1-Score', fontsize=12)
axes[2].set_ylabel('Obesity Class', fontsize=12)
axes[2].set_title('F1-Score by Class (Cross-Validated)', fontsize=14, fontweight='bold')
axes[2].set_xlim([0, 1])
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==========================================
# Part 6: Train Final Model (Using Full Dataset)
# ==========================================
print(f"\n{'='*60}")
print("Training Final Model on Full Dataset")
print(f"{'='*60}")

# Use average best iteration from CV
avg_best_iter = int(scores_df['best_iteration'].mean())
print(f"\nUsing average best iteration: {avg_best_iter}")

full_pool = Pool(
    data=X,
    label=y,
    cat_features=cat_feature_indices
)

final_model = CatBoostClassifier(
    loss_function="MultiClass",
    eval_metric="TotalF1",
    learning_rate=BEST_LR,
    depth=BEST_DEPTH,
    l2_leaf_reg=BEST_L2_REG,
    random_seed=RANDOM_STATE,
    iterations=avg_best_iter,
    verbose=100
)

final_model.fit(full_pool)

# Save model
final_model.save_model("catboost_obesity_model_cv.cbm")
print("\nFinal CatBoost model saved to catboost_obesity_model_cv.cbm")

# ==========================================
# Part 7: Test Set Predictions (Ensemble All Fold Models)
# ==========================================
print(f"\n{'='*60}")
print("Making Test Predictions (Ensemble)")
print(f"{'='*60}")

# Method 1: Average probabilities from all fold models
test_proba_avg = np.mean(test_predictions, axis=0)
test_pred_ensemble = fold_models[0].classes_[np.argmax(test_proba_avg, axis=1)]

# Method 2: Use final model predictions
test_pool_final = Pool(
    data=test_df,
    cat_features=cat_feature_indices
)
test_pred_final = final_model.predict(test_pool_final).reshape(-1)

# ==========================================
# Part 8: Generate Submission Files
# ==========================================
# Using ensemble predictions
submission_ensemble = sample_submission.copy()
submission_ensemble[TARGET_COL] = test_pred_ensemble
submission_ensemble.to_csv("catboost_submission_cv_ensemble.csv", index=False)

# Using final model predictions
submission_final = sample_submission.copy()
submission_final[TARGET_COL] = test_pred_final
submission_final.to_csv("catboost_submission_cv_final.csv", index=False)

print("\n" + "="*60)
print("Submission files have been created:")
print("  1. catboost_submission_cv_ensemble.csv (CV ensemble)")
print("  2. catboost_submission_cv_final.csv (final model)")
print("="*60)

print("\nEnsemble submission preview:")
print(submission_ensemble.head(10))

print("\nFinal model submission preview:")
print(submission_final.head(10))
```
